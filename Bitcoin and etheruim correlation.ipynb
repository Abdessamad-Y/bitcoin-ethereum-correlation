```python
bitcoin.loc['2020','Close'].resample('M').mean().plot()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2020','Close'].plot()
bitcoin.loc['2020','Close'].resample('M').mean().plot(label='moyenne par mois',lw=3,ls=':',alpha=0.8)
bitcoin.loc['2020','Close'].resample('W').mean().plot(label='moyenne par semaine',lw=3,ls=':',alpha=0.8)
plt.legend()
plt.show()
```
